```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data  = pd.read_csv('all_data.csv')
print(data.head())
```

```python
# This graph will vizualize life expectancy over time for different countries(6) that have been provided in the dataset.
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Year', y='Life expectancy at birth (years)', hue='Country', marker='o')
plt.title('Life Expectancy Over Time by Country')
plt.xlabel('Year')
plt.ylabel('Life Expectancy at Birth (years)')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
plt.clf()
```

```python
# As we can see from the graph, life expectancy has generally increased over time for all countries in the dataset. However, the rate of increase varies between countries, with some showing more significant improvements than others. This could be due to various factors such as healthcare advancements, economic development, and social policies. Further analysis would be needed to understand the specific reasons behind these trends.
```

```python
#This graph will visualize GDP over time for different countries(6) that have been provided in the dataset. The code is almost identical to the previous graph, with only the y-axis variable changed from life expectancy to GDP.
plt.figure(figsize=(12, 6))
sns.lineplot(data=data, x='Year', y='GDP', hue='Country', marker='o')
plt.title('GDP Over Time by Country')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
plt.clf()
```

```python
#This graph will answer the question: Is there a correlation between GDP and life expectancy?
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='GDP', y='Life expectancy at birth (years)', hue='Country', style='Country', s=100)
plt.title('Correlation between GDP and Life Expectancy')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy at Birth (years)')
plt.legend(title='Country', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
plt.clf()
```

```python
#From this single graph, we can notice quite a few trends. 
#1. We notice that both China and the united states are trending further rigth on the x-axis, indicating that their GDP is increasing over time quicker than the other countries.
#2. We can also see that as GDP increases, life expectancy tends to increase as well. This suggests a positive correlation between GDP and life expectancy, indicating that countries with higher GDPs tend to have higher life expectancies.
#3. Interestingly, we notice that Zimbabwe has, over time, not grown their GDP, respectively, nearly as much as the GDP leaders in the Dataset, however- their life expectancy has still increased significantly. This could suggest that factors other than GDP are also playing a significant role in improving life expectancy in Zimbabwe.
```

```python

avg_life_expectancy = data.groupby('Country')['Life expectancy at birth (years)'].mean().reset_index()
avg_life_expectancy = avg_life_expectancy.sort_values('Life expectancy at birth (years)', ascending=False)
print("Average Life Expectancy by Country (2000-2015):")
print(avg_life_expectancy)
print()

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_life_expectancy, x='Country', y='Life expectancy at birth (years)', palette='viridis')
plt.title('Average Life Expectancy by Country (2000-2015)', fontsize=14, fontweight='bold')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Average Life Expectancy (years)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
plt.clf()
```
